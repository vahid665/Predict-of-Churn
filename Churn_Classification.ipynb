```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['state'], df['churn'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
```

```python
chi2, p, dof, expected = chi2_contingency(contingency_table)

n = contingency_table.sum().sum()
min_dim = min(contingency_table.shape) - 1 
cramers_v = np.sqrt(chi2 / (n * min_dim))

print(f"Cramér's V: {cramers_v}")
```

```python
# Cramér's V Dəyəri	Əlaqə Gücü
# 0 - 0.1	Zəif əlaqə
# 0.1 - 0.3	Orta əlaqə
# 0.3 - 0.5	Güclü əlaqə
# 0.5+	Çox güclü əlaqə
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
y = df['churn']
X = df.drop(axis=1,columns='churn')
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, train_size=0.8)
```

```python
frequency_encoding = X_train['state'].value_counts().to_dict()

X_train['state'] = X_train['state'].map(frequency_encoding)
X_test['state'] = X_test['state'].map(frequency_encoding)
X_test['state'].fillna(X_train['state'].median(), inplace=True)
```

```python
df.head()
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse_output=False)  # Sparse matrix qaytarmasın
X_train_encoded = encoder.fit_transform(X_train[['area_code']])
X_test_encoded = encoder.transform(X_test[['area_code']])

encoded_columns = encoder.get_feature_names_out(['area_code'])
X_train_encoded_df = pd.DataFrame(X_train_encoded, columns=encoded_columns, index=X_train.index)
X_test_encoded_df = pd.DataFrame(X_test_encoded, columns=encoded_columns, index=X_test.index)

# Orijinal DataFrame ilə birləşdirmək
X_train = pd.concat([X_train.drop(columns=['area_code']), X_train_encoded_df], axis=1)
X_test = pd.concat([X_test.drop(columns=['area_code']), X_test_encoded_df], axis=1)
```

```python
X_train.head()
```

```python
X_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for col in ['voice_mail_plan', 'international_plan']:
    X_train[col] = encoder.fit_transform(X_train[col])
    X_test[col] = encoder.transform(X_test[col])
```

```python
X_train.info()
```

# Model RandomForest

```python
model_random_forest = RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
model_random_forest.fit(X_train,y_train)
```

```python
model_random_forest.score(X_test,y_test)
```
